```python
pattern=re.compile(r"\b(?:(192.168.(25[0-5]).([0-9]|[1-2][0-4][0-9]|25[0-4]))$)\b")
raw="192.168.253.23, 192.168.253.212 192.168.200.95 192.168.255.24"
x=re.findall(pattern,raw)
d=[]
for i in x:
    d.append(i[0])
print(d)
```
