```python
def create_regex_for_ip(ip_with_range):
    ip=ip_with_range.split(',')
    data=pd.DataFrame()
    for i in ip:
        ip_subnet=i.rfind('.')
#         print(ip_subnet)
        slash=i.rfind('/')
        subnet_end=int(i[(slash)+1:])
        if subnet_end==24:
            pattern='''^({}.(?:[1-9]|[1-9][0-9]|[1][0-9][0-9]|[2][0-4][0-9]|25[0-4]))$'''.format((i[:ip_subnet]).strip())
            dataframe=[{
                "Department_Name":"Python",
                "End_IP":i[:ip_subnet]+'.254',
                "IP_Subnet":i,
                "Start_IP":i[:ip_subnet]+'.0',
                "CustomerName":"SATTRIX",
                "regex":pattern
            }]
            ip_rows=pd.DataFrame(dataframe)
            data=pd.concat([ip_rows,data],ignore_index=True)
        if subnet_end==25:
            pattern='''^({}.(?:12[8-9]|1[3-9][0-9]|2[0-4][0-9]|25[0-4]))$'''.format(i[:ip_subnet])
            dataframe=[
                    {
                        "Department_Name":"Python",
                        "End_IP":i[:ip_subnet]+'.254',
                        "IP_Subnet":i,
                        "Start_IP":i[:ip_subnet]+'.128',
                        "CustomerName":"SATTRIX",
                        "regex":pattern
                    }
                ]
            ip_rows=pd.DataFrame(dataframe)
            data=pd.concat([ip_rows,data],ignore_index=True)
        if subnet_end==26:
            pattern='''^({}.(?:19[2-9]|2[0-4][0-9]|25[0-4]))$'''.format(i[:ip_subnet])
            dataframe=[
                    {
                        "Department_Name":"Python",
                        "End_IP":i[:ip_subnet]+'.254',
                        "IP_Subnet":i,
                        "Start_IP":i[:ip_subnet]+'.192',
                        "CustomerName":"SATTRIX",
                        "regex":pattern
                    }
                ]
            ip_rows=pd.DataFrame(dataframe)
            data=pd.concat([ip_rows,data],ignore_index=True)
        if subnet_end==27:
            pattern='''^({}.(?:22[4-9]|2[3-4][0-9]|25[0-4]))$'''.format(i[:ip_subnet])
            dataframe=[
                    {
                        "Department_Name":"Python",
                        "End_IP":i[:ip_subnet]+'.254',
                        "IP_Subnet":i,
                        "Start_IP":i[:ip_subnet]+'.224',
                        "CustomerName":"SATTRIX",
                        "regex":pattern
                    }
                ]
            ip_rows=pd.DataFrame(dataframe)
            data=pd.concat([ip_rows,data],ignore_index=True)
        if subnet_end==28:
            pattern='''^({}.(?:[2][4][0-9]|25[0-4]))$'''.format(i[:ip_subnet])
            dataframe=[
                    {
                        "Department_Name":"Python",
                        "End_IP":i[:ip_subnet]+'.254',
                        "IP_Subnet":i,
                        "Start_IP":i[:ip_subnet]+'.240',
                        "CustomerName":"SATTRIX",
                        "regex":pattern
                    }
                ]
            ip_rows=pd.DataFrame(dataframe)
            data=pd.concat([ip_rows,data],ignore_index=True)
        if subnet_end==29:
            pattern='''^({}.(?:[2][4][8-9]|25[0-4]))$'''.format(i[:ip_subnet])
            dataframe=[
                    {
                        "Department_Name":"Python",
                        "End_IP":i[:ip_subnet]+'.254',
                        "IP_Subnet":i,
                        "Start_IP":i[:ip_subnet]+'.248',
                        "CustomerName":"SATTRIX",
                        "regex":pattern
                    }
                ]
            ip_rows=pd.DataFrame(dataframe)
            data=pd.concat([ip_rows,data],ignore_index=True)
        if subnet_end==30:
            pattern='''^({}.(?:25[2-4]))$'''.format(i[:ip_subnet])
            dataframe=[
                    {
                        "Department_Name":"Python",
                        "End_IP":i[:ip_subnet]+'.254',
                        "IP_Subnet":i,
                        "Start_IP":i[:ip_subnet]+'.252',
                        "CustomerName":"SATTRIX",
                        "regex":pattern
                    }
                ]
            ip_rows=pd.DataFrame(dataframe)
            data=pd.concat([ip_rows,data] , ignore_index=True)
    return data
```
